```python
#Problem 1 part 1
import numpy as np
x = np.linspace(-2,2,1000)
y = np.linspace(-2,2,1000)
X,Y = np.meshgrid(x,y)

z = 0
C = X + Y*1j
for i in range(100):
    z = z**2 + C
div = np.isnan(z)

import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 7))

    
ax.imshow(div, extent=[-2,2,-2,2],cmap = 'plasma',vmin=0,vmax=1)
```

```python
#Problem 1 part 2
iteration_num = np.zeros_like(C).astype(int)
x = np.linspace(-2,2,1000)
y = np.linspace(-2,2,1000)
X,Y = np.meshgrid(x,y)

z = 0
C = X + Y*1j
for i in range(100):
    z = z**2 + C
    iteration_num += np.isfinite(z)
    
'''setting the converging coordinates as negative iteration number to 
obtain contrast in the plot
'''

iteration_num = np.where(iteration_num == 100,-20,iteration_num)    
    
import matplotlib.pyplot as plt


plt.figure(dpi=100)
plt.contourf(X,Y,iteration_num,levels=50)
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label="iteration number",)
plt.show()

        
    
```

```python
#Problem 2
import scipy
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def SIR(y,t,b,g):
    S, I, R = y
    dS_dt = -b*S*I/1000
    dI_dt = b*S*I/1000-g*I
    dR_dt = g*I
    
    return(dS_dt,dI_dt,dR_dt)
```

```python
#Initial Conditions
I0 = 1
S0 = 999
R0 = 0
```

```python
#parameters 1
b = 2
g = 2
#time
t = np.linspace(0,200,10000)

solution = scipy.integrate.odeint(SIR, [S0,I0,R0],t,args=(b,g))
output = np.array(solution)
plt.figure(figsize=[8,5])
plt.title("SIR Model")
plt.xlabel('Time (days)')
plt.ylabel('Number of People')
plt.plot(t,output[:,0],label='S(t)',color = 'black')
plt.plot(t,output[:,1],label='I(t)',color = 'green')
plt.plot(t,output[:,2],label='R(t)',color = 'blue')
plt.legend()
```
